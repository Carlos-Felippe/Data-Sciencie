```
#Grafico com ligação de pontos

import matplotlib.pyplot as plt

x =[1 ,2 ,3 ,4 ,5 ,6]
y =[2 ,4,5 ,7 ,-3 ,9]

titulo='Scatterplot: gráfico de dispersão'
eixox="Eixo X"
eixoy="Eixo Y"

#legendas
plt.title(titulo, size="18")
plt.xlabel(eixox)
plt.ylabel(eixoy, color="b")

# s é atributo para aumentar o marker
plt.scatter(x, y, label="Meus pontos", color="#FF00FF", marker="h", s=200) 

# serve para ligar os pontos exitesntes no gráfico linestyle e usando outras conf.
plt.plot(x, y, label="ligação dos pontos", linestyle="--")

#exibe as label
plt.legend()
plt.show()
```
