```
#Grafico com ligação de pontos com dimensão em -- Z --

import matplotlib.pyplot as plt

x =[1 ,2 ,3 ,4 ,5 ,6]
y =[2 ,4, 5 ,7 ,-3 ,9]
z =[30, 80, 400, 450, 600, 500]# valor dos pontos da dimensão Z

titulo='Scatterplot: gráfico de dispersão em 3 dimensão'
eixox="Eixo X"
eixoy="Eixo Y"

#legendas
plt.title(titulo, size="18")
plt.xlabel(eixox)
plt.ylabel(eixoy, color="b")

# s agora pode ser outra dimensão
plt.scatter(x, y, label="Meus pontos", color="k", marker=".", s=z) 

# serve para ligar os pontos exitesntes no gráfico linestyle e usando outras conf.
plt.plot(x, y, label="ligação dos pontos", linestyle="--")

#exibe as label
plt.legend()
plt.show()
```
