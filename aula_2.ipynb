```
#Visualização de dados em Python

import matplotlib.pyplot as plt

x = [1, 2, 4, 5]
y = [2, 3 ,8, 9]

#titulo

plt.title("Meu primeiro gráfico em Python" , color="red", size= 18)

#Eixos

plt.xlabel("Eixo X", size=16)
plt.ylabel("Eixo Y", size=16)




plt.plot(x, y)
plt.show()
```
